**Inference**: So we have detected one outlier here. I grouped data like that because if we simply find outliers of 'Value' it might not be much meaningful.ey
Even though we have outliers here i assume we cannot consider them as error or impossible values. They might be genuine values, so **i am using the Winsorization method to treat outliers here**, and **creating a new feature which contains the Winsorzed values of the 'Value' column**. **So i am keeping this feature  apart and continuing with 'Value' column itself for the further procedure since we are not sure those are ingenuine values**.

```python
from scipy.stats.mstats import winsorize

# Apply winsorization to 'Value' column
data['Winsorized_Value'] = winsorize(data['Value'], limits=[0.05, 0.05])

# Check the box plot before and after winsorization
data.boxplot(column=['Value', 'Winsorized_Value'])
```

Inference: Here we can see that after Winsorization, the outliers in the Value column have decreased significantly which indicates that our method was successful.
